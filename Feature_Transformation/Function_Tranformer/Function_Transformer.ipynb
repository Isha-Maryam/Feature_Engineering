```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
```

```python
df=pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,1:3]
y=df.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['Age'])
plt.title("Age PDF")
plt.subplot(122)
stats.probplot(X_train['Age'],dist='norm',plot=plt)
plt.title("Age QQ Plot")
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['Fare'])
plt.title("Fare PDF")
plt.subplot(122)
stats.probplot(X_train['Fare'],dist='norm',plot=plt)
plt.title("Fare QQ Plot")
plt.show()
```

```python
clf=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
clf2.fit(X_train,y_train)
y_pred=clf.predict(X_test)
y_pred1=clf2.predict(X_test)
```

Accuracy without Function Transformer

```python
print("Accuracy LR: ",accuracy_score(y_test,y_pred))
print("Accuracy DT: ",accuracy_score(y_test,y_pred1))
```

```python
trf=FunctionTransformer(func=np.log1p)
```

```python
X_train_transformed=trf.fit_transform(X_train)
X_test_transformed=trf.transform(X_test)
```

```python
clf=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf.fit(X_train_transformed,y_train)
clf2.fit(X_train_transformed,y_train)
y_pred2=clf.predict(X_test_transformed)
y_pred3=clf2.predict(X_test_transformed)
```

Improved Accuracy after Function Transformer

```python
print("Accuracy of LR After FunctionTransform: ",accuracy_score(y_test,y_pred2))
print("Accuracy of DT After FunctionTransform: ",accuracy_score(y_test,y_pred3))
```

To check the accuracy  changes  we will cross validate it

```python
X_transformed=trf.fit_transform(x)
clf=LogisticRegression()
clf2=DecisionTreeClassifier()
print("LR",np.mean(cross_val_score(clf,X_transformed,y,scoring="accuracy",cv=10)))
print("DT",np.mean(cross_val_score(clf2,X_transformed,y,scoring="accuracy",cv=10)))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(X_train['Fare'],dist="norm",plot=plt)
plt.title("Fare Before log")
plt.subplot(122)
stats.probplot(X_train_transformed['Fare'],dist="norm",plot=plt)
plt.title("Fare After log")
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
stats.probplot(X_train['Age'],dist="norm",plot=plt)
plt.title("Age Before log")
plt.subplot(122)
stats.probplot(X_train_transformed['Age'],dist="norm",plot=plt)
plt.title("Age After log")
plt.show()
```

Customize Fucntion to check 'Fare' accuracy by every method (square, square root, reciprocal, sin etc)

```python
def apply_transform(transform_func):
    X = df.iloc[:,1:3]
    y = df.iloc[:,0]

    trf = ColumnTransformer([('log', FunctionTransformer(transform_func), ['Fare'])], remainder='passthrough')

    X_trans = trf.fit_transform(X)

    clf = LogisticRegression()

    print("Accuracy", np.mean(cross_val_score(clf, X_trans, y, scoring='accuracy', cv=10)))

    plt.figure(figsize=(14,4))

    plt.subplot(121)
    stats.probplot(X['Fare'], dist="norm", plot=plt)
    plt.title('Fare Before Transform')

    plt.subplot(122)
    stats.probplot(X_trans[:,0], dist="norm", plot=plt)
    plt.title('Fare After Transform')

    plt.show()
```

```python
apply_transform(np.log1p)
```
